```python
input_seq_iterator = SeqIO.parse(genbank_file, "gb")
with open("Gene_list_from_genbank_v2.tsv", 'w+') as f:
    for seq_record in input_seq_iterator:
            #print("%s, %s..." % (seq_record.id, seq_record.description))
            #print("Sequence length %i, %i features, from: %s \n\n"
            #      % (len(seq_record), len(seq_record.features), seq_record.annotations["source"]))

            for feature in seq_record.features:
                if (feature.type == "CDS"):
                    #Set start to + or -
                    if feature.strand == 1:
                        strand = "+"
                    elif feature.strand == -1:
                        strand = "-"
                    #Set locus
                    locus = feature.qualifiers.get('locus_tag')[0]
                    #Set gene name, if not use locus name
                    if feature.qualifiers.get('gene'):
                        gene = feature.qualifiers.get('gene')[0]
                    else:
                        gene = locus
                    #Set start and and
                    start = str(feature.location.start)
                    end = str(feature.location.end)
                    line = ("\t").join([gene, locus, start, end, strand]) + "\n"
                    print(line)
                    f.write(line)
```
